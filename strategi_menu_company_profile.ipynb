## 1. Prinsip Strategis Penentuan Menu
Untuk menghasilkan website company profile yang menarik, profesional, dan mencerminkan kapabilitas NJB sebagai galangan kapal terintegrasi, penentuan menu harus mengikuti prinsip:

- **Clarity (Kejelasan)**: Pengunjung dapat memahami struktur informasi dalam 3 detik.
- **Hierarchy (Hirarki)**: Informasi terpenting muncul lebih dulu.
- **Credibility (Kepercayaan)**: Menonjolkan bukti kemampuan: fasilitas, proyek, sertifikasi, dan proses engineering.
- **Conversion-Oriented**: Mendorong pengunjung ke aksi: inquiry, form quotation, atau contact.


## Struktur Menu yang Direkomendasikan

### **Beranda (Home)**
Isi:
- Tagline: *“Built to Last – Engineering Maritime Excellence”*
- Hero section: Foto fasilitas, kapal, atau proses fabrikasi
- Highlight cepat: Kapabilitas, tahun pengalaman, jumlah proyek
- Tombol CTA: Inquiry / Hubungi Kami

---

### **Tentang Kami (About Us)**
Isi:
- Corporate Synopsis
- Sejarah perusahaan
- Visi & Misi
- Nilai & Budaya Kerja
- Struktur organisasi ringkas

---

### **Layanan & Kapabilitas (Capabilities)**
Isi:
- New Building (Deck Cargo Barge, Tugboat, SPOB, Oil Barge)
- Marine Structure Fabrication
- Repair & Maintenance (opsional, bila ingin ditambahkan)
- Kekuatan engineering (Design–Build–Maintain)

---

### **Fasilitas Produksi (Infrastructure)**
Isi:
- Heavy Lifting Equipment
- Fabrication Workshops
- Machine Shop
- Utility Systems
- Launching Facility
- Foto dan ilustrasi tata letak shipyard

---

### **Proyek & Portofolio (Projects)**
Isi:
- Showcase kapal-kapal berhasil dibuat

---

### **Sertifikasi & HSE (ISO & Safety)**
Isi:

- Proses QC & QC Plan

### **Kontak & Inquiry (Contact)**
